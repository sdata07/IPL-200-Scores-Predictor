```python
def gradient(x,y,w,b) :
    dj_dw = (1 / len(x)) * np.sum((w * x + b - y) * x)
    dj_db = (1 / len(x)) * np.sum(w * x + b - y)
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w, b, alpha, num_iters) : 

    
    for i in range(num_iters) :
        dj_dw, dj_db = gradient(x, y, w, b)
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

    return w,b
```
