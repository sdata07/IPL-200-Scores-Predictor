```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math,copy

data = pd.read_csv("./ipl_200plus_scores.csv")
data.head(20)

years = data["Season"] - 2008
scores = data["200+ Scores"]
```

```python
def cost_count(x,y,w,b) :
    return 1/(2 * len(x)) * np.sum((y - (x * w + b))**2)
```

```python
def gradient(x,y,w,b) :
    dj_dw = (1 / len(x)) * np.sum((w * x + b - y) * x)
    dj_db = (1 / len(x)) * np.sum(w * x + b - y)
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w, b, alpha, num_iters) : 

    
    for i in range(num_iters) :
        dj_dw, dj_db = gradient(x, y, w, b)
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

    return w,b
```

```python


plt.scatter(years, scores)
plt.xlabel("Year")
plt.ylabel("No. of 200+ scores")

w,c = gradient_descent(years,scores, 2,3, 0.01, 10000)
print("weight (w) : ", w)
print("bias (c) : ", c)

cost = cost_count(years,scores, w, c)

print("cost : ", cost) 
plt.plot(years, w * years + c, color="Orange")
plt.show()
```
